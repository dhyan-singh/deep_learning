```python
import os, sys, subprocess
from pathlib import Path

REPO_URL = "https://github.com/dhyan-singh/deep_learning.git"
REPO_DIR = "/content/deep_learning"

# Clone the repository if it doesn't already exist
if not Path(REPO_DIR).exists():
    subprocess.check_call(["git", "clone", REPO_URL, REPO_DIR])

# Change the current working directory to the repository
os.chdir(REPO_DIR)

# Install the dl_utils package in editable mode quietly
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "-e", "packages/dl_utils"])

# Add the 'packages' directory to sys.path if not already present
packages_path = os.path.join(REPO_DIR, 'packages')
if packages_path not in sys.path:
    sys.path.insert(0, packages_path)

# Import get_device from the installed package and print the detected device
from dl_utils.dl_utils.device import get_device
print(f"Using device: {get_device()}")
```
